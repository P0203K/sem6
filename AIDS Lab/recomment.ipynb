```python
# Importing necessary libraries
import pandas as pd
from sklearn.neighbors import NearestNeighbors
```

```python
# Load the dataset
songs_data = pd.read_csv('data.csv')
songs_data.head()
```

```python
songs_data.shape
```

```python
songs_data.isnull().sum()
```

```python
# Dropping non-numeric columns if any
songs_data_numeric = songs_data.select_dtypes(include='number')
songs_data_numeric.columns
```

```python
# Create a NearestNeighbors model
knn_model = NearestNeighbors(n_neighbors=5, algorithm='auto')
```

```python
# Fit the model with song attributes
knn_model.fit(songs_data_numeric)
```

```python
# Function to recommend songs
def recommend_songs(song_name):
    try:
        # Finding the index of the given song
        song_index = songs_data[songs_data['name'] == song_name].index[0]
        distances, indices = knn_model.kneighbors(songs_data_numeric.iloc[song_index, :].values.reshape(1, -1))
        print("Recommended Songs:")
        for i in range(1, len(indices.flatten())):
            print(f"{i}: {songs_data['name'][indices.flatten()[i]]}")
    except IndexError:
        print("Song not found in the dataset. Please try another song.")
```

```python
# Taking user input for the song name
user_input = input("Enter a song name: ")
recommend_songs(user_input)
```
