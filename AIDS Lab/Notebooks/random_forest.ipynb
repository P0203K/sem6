```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_xlsx(r'C:\Users\Preeti Khamkar\OneDrive\Desktop\AIDS Lab\salary.xlsx')
df.dropna(inplace=True)
df.head()
```

```python
x = df[['Temperature (C)', 'Apparent_Temperature_C', 'Humidity', 'Visibility (km)', 'Pressure (millibars)']]
y = df['Summary']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=42)
X_train.shape, X_test.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier_rf = RandomForestClassifier(random_state=42, n_jobs=-1, max_depth=5,
                                       n_estimators=100, oob_score=True)
classifier_rf.fit(X_train, y_train)
```

```python
classifier_rf.oob_score_
```

```python
y_pred = classifier_rf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy: ", accuracy)
df = pd.DataFrame(y_test, y_pred)
df = pd.DataFrame({"Actual Summary":y_test, "Predicted Summary":y_pred})
df
```

```python
classification_rep = classification_report(y_test, y_pred)
print("Report:",classification_rep)
```
