```python
data = df[['Precip_Type', 'Humidity', 'Temperature (C)']]
```

```python
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
```

```python
data = data.replace('rain',0)
data = data.replace('snow', 1)
```

```python
from sklearn.model_selection import GridSearchCV
parameter ={'penalty' : ['l2'] , 'C': [1,2,3,4,5,6,10,20,30,40,50] , 'max_iter': [100,200,300,]}
```

```python
# from sklearn.model_selection import train_test_split
x = data[['Temperature (C)', 'Humidity']]
y = data['Precip_Type']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
reg_model = GridSearchCV(reg , param_grid = parameter , scoring = 'accuracy' , cv= 5 , error_score='raise')
reg_model.fit(x_train , y_train)
```
